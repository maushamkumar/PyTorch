# Creating Tensor
 A torch.Tensor is a multi-dimensional matrix containing elements of a single data type.

 A tensor can be constructed from a python list or sequence using the torch.tensor() constructor.  

 ## Tensor class reference
There are a few ways to create a tensor, depending on your use case.
- To create a tensor with pre-existing data, use torch.tensor().
- To create a tensor with specific size, use torch.* tensor creation ops

Note:- you might've noticed me using lowercase letters for scalar and vector and uppercase letters for MATRIX and TENSOR. This was on purpose. In practice, you'll often see scalar and vector denoted as lowercase such as y or a. And matrices and tensors denoted as uppercase letter such as X or A.

you also might notice the names matrix and tensor used interchangeably. This is common. since in PyTorch you're often dealing with torch.Tensor (hence the tensor name), however, the shape and dimensions of what's inside will dictate what it actually is.

## Random tensor

Let's see how to create a tensor of random numbers.
we can do so using torch.rand() adn passing in the size parameter.

```python
# Create a random tensor of size (3, 4
random_tensor = torch.rand(3, 4)
random_tensor, random_tensor.dtype
```

```python
# Create a random tensor of size (224, 224, 3)
random_image_tensor = torch.rand(size=(224, 224, 3))
random_image_tensor.shape, random_image_tensor.ndim
```

## Zeros and Ones

Let's Create a tensor full of zeros with torch.zeros()

```python
# Create a tensor of all zeros
zeros = torch.zeros(3, 4)
zeros, zeros.dtype
```

we can do the same to create a tensor of all ones except using torch.ones() instead.

```python
ones = torch.ones(3,5)
ones, ones.dtype
```

```python
zeros1 = torch.zeros(size=(3, 4), dtype=torch.float16)
zeros1, zeros1.dtype
```

```python
ones1 = torch.ones(size=(3, 4), dtype=torch.float16)
ones1, ones1.dtype
```
